Q1. 95% Confidence Interval using Python

```python
import scipy.stats as stats
import numpy as np

mean = 50
std = 5
n = 30
confidence = 0.95

z = stats.norm.ppf((1 + confidence) / 2)

margin_error = z * (std / np.sqrt(n))

lower = mean - margin_error
upper = mean + margin_error

print("Confidence Interval:", (lower, upper))
```

Q2. Chi-square Goodness of Fit Test (M&M Colors)

```python
import scipy.stats as stats

observed = [18, 22, 19, 9, 11, 21]   # example observed counts
expected_ratio = [0.2,0.2,0.2,0.1,0.1,0.2]

total = sum(observed)
expected = [r*total for r in expected_ratio]

chi2, p = stats.chisquare(observed, expected)

print("Chi-square:", chi2)
print("p-value:", p)
```

Q4. 95% Confidence Interval for Smoking Proportion

```python
import numpy as np
import scipy.stats as stats

n = 500
smokers = 60

p = smokers/n
z = stats.norm.ppf(0.975)

margin = z * np.sqrt((p*(1-p))/n)

lower = p - margin
upper = p + margin

print("CI:", (lower, upper))
```

Q5. 90% Confidence Interval

```python
import scipy.stats as stats
import numpy as np

mean = 75
std = 12
n = 30

z = stats.norm.ppf(0.95)

margin = z*(std/np.sqrt(n))

lower = mean - margin
upper = mean + margin

print("CI:", (lower, upper))
```

Q6. Plot Chi-square Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df = 10

x = np.linspace(0,30,500)
y = chi2.pdf(x,df)

plt.plot(x,y)

x_fill = np.linspace(15,30,200)
plt.fill_between(x_fill,chi2.pdf(x_fill,df))

plt.xlabel("Chi-square value")
plt.ylabel("Probability Density")
plt.title("Chi-square Distribution (df=10)")

plt.show()
```

Q7. 99% Confidence Interval for Coke Preference

```python
import scipy.stats as stats
import numpy as np

n = 1000
coke = 520

p = coke/n
z = stats.norm.ppf(0.995)

margin = z*np.sqrt((p*(1-p))/n)

lower = p-margin
upper = p+margin

print("99% CI:",(lower,upper))
```

Q11. Hypothesis Test for Population Mean

```python
import scipy.stats as stats
import numpy as np

mean = 72
mu = 70
std = 10
n = 30

t = (mean-mu)/(std/np.sqrt(n))

p = stats.t.sf(abs(t),df=n-1)*2

print("t-statistic:",t)
print("p-value:",p)
```

Interpretation

p < 0.05 → Reject null hypothesis

p > 0.05 → Mean is not significantly different from 70
